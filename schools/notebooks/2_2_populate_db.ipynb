```python
from utils import MysqlConnector, FakeData
```

```python
conn = MysqlConnector()
```

```python
print('Checking rows in each table...')

tables = ['lessons','dates','schools','school_subjects','teachers','class_types','classrooms','classes']
for table in tables:
    rows = conn.fetch(f"SELECT COUNT(*) FROM {table}")
    print(f"{table}: {rows[0][0]}")
```

```python
if conn.fetch("SELECT COUNT(*) FROM dates")[0][0] == 0:
    print('Populating dates table...')
    dates = FakeData().dates_df
    dates_values = dates.values.tolist()
    dates_columns = dates.columns.tolist()
    conn.batch_insert('dates', dates_columns, dates_values)
else:
    print('Dates table already populated')
```

```python
if conn.fetch("SELECT COUNT(*) FROM school_subjects")[0][0] == 0:
    print('Populating school_subjects table...')
    school_subjects = FakeData().school_subjects_df
    school_subjects_values = school_subjects.values.tolist()
    school_subjects_columns = school_subjects.columns.tolist()
    conn.batch_insert('school_subjects', school_subjects_columns, school_subjects_values)
else:
    print('School_subjects table already populated')
```

```python
if conn.fetch("SELECT COUNT(*) FROM class_types")[0][0] == 0:
    print('Populating class_types table...')
    class_types = FakeData().class_types_df
    class_types_values = class_types.values.tolist()
    class_types_columns = class_types.columns.tolist()
    conn.batch_insert('class_types', class_types_columns, class_types_values)
else:
    print('Class_types table already populated')
```

```python
def insert_schools(number_of_schools):
    fake_obj = FakeData(number_of_schools=number_of_schools)
    tables = [
        ('schools',fake_obj.schools_df),
        ('teachers',fake_obj.teachers_df),
        ('classrooms',fake_obj.classrooms_df),
        ('classes',fake_obj.classes_df),
        ('lessons',fake_obj.lessons_df)
    ]
    for table, data in tables:
        print(f'Populating {table} table...')
        values = data.values.tolist()
        columns = data.columns.tolist()
        conn.batch_insert(table, columns, values)
```

```python
for i in range(18):
    print(f'Inserting school {i+1}...')
    insert_schools(1)
    print('Done')
```
